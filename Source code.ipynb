# ***D. Performance Aggregation and Metrics Calculation***
1. Initialize performance_by_topic
2. Populate performance_by_topic with all relevant values
3. Add calculated metrics to performance_by_topic
4. Final Output

```python
# Historical quiz data
historical_quiz_data = {
    336566: {
        'id': 336566,
        'quiz_id': 43,
        'user_id': '7ZXdz3zHuNcdg9agb5YpaOGLQqw2',
        'submitted_at': '2025-01-17T15:51:29.859+05:30',
        'score': 32,
        'correct_answers': 8,
        'incorrect_answers': 2,
        'negative_score': 2.0,
        'quiz': {
            'id': 43,
            'topic': 'Structural Organisation in Animals',
        },
    },
    336567: {
        'id': 336567,
        'quiz_id': 44,
        'user_id': '9ZXdz3zHuNcdg9agb5YpaOGLQqw2',
        'submitted_at': '2025-01-18T15:51:29.859+05:30',
        'score': 25,
        'correct_answers': 5,
        'incorrect_answers': 3,
        'negative_score': 1.5,
        'quiz': {
            'id': 44,
            'topic': 'Cell Biology',
        },
    },
}

performance_by_topic = {}


for key, quiz_data in historical_quiz_data.items():
    topic = quiz_data['quiz']['topic']
    if topic not in performance_by_topic:
        performance_by_topic[topic] = {
            'total_quizzes': 0,
            'correct_answers': 0,
            'incorrect_answers': 0,
            'score': 0,
            'negative_score': 0,
        }
    performance_by_topic[topic]['total_quizzes'] += 1
    performance_by_topic[topic]['correct_answers'] += quiz_data['correct_answers']
    performance_by_topic[topic]['incorrect_answers'] += quiz_data['incorrect_answers']
    performance_by_topic[topic]['score'] += quiz_data['score']
    performance_by_topic[topic]['negative_score'] += quiz_data['negative_score']


for topic, performance in performance_by_topic.items():
    total_questions = performance['correct_answers'] + performance['incorrect_answers']
    accuracy = (performance['correct_answers'] / total_questions) * 100 if total_questions > 0 else 0
    average_score = performance['score'] / total_questions if total_questions > 0 else 0
    average_negative_score = performance['negative_score'] / performance['total_quizzes'] if performance['total_quizzes'] > 0 else 0
    performance['accuracy'] = round(accuracy, 2)
    performance['average_score'] = round(average_score, 2)
    performance['average_negative_score'] = round(average_negative_score, 2)


print("Performance by Topic:")
for topic, performance in performance_by_topic.items():
    print(f"Topic: {topic}")
    print(f"  Total Quizzes: {performance['total_quizzes']}")
    print(f"  Correct Answers: {performance['correct_answers']}")
    print(f"  Incorrect Answers: {performance['incorrect_answers']}")
    print(f"  Total Score: {performance['score']}")
    print(f"  Total Negative Score: {performance['negative_score']}")
    print(f"  Accuracy: {performance['accuracy']}%")
    print(f"  Average Score per Question: {performance['average_score']}")
    print(f"  Average Negative Score per Quiz: {performance['average_negative_score']}")
    print()
```

# ***F. Provide Personalized Recommendations Based on Performance Data***
1.  Sample DataFrame with quiz performance data
2. Initialize categories for feedback
3. Define the thresholds
4. Iterate over the DataFrame rows to give feedback
5. Output feedback

```python
data = {
    'Topic': ['Structural Organisation in Animals', 'Cell Biology'],
    'total_quizzes': [1, 1],
    'correct_answers': [8, 5],
    'incorrect_answers': [2, 3],
    'score': [32, 25],
    'negative_score': [2.0, 1.5],
    'accuracy': [80.0, 62.5],
    'average_score': [3.20, 3.12],
    'average_negative_score': [2.0, 1.5],
    'total_score': [30.0, 24.0],
}

df_performance = pd.DataFrame(data)


feedback = {
    "Critical Improvement": [],
    "Moderate Improvement": [],
    "Good Performance": [],
    "Mistake Reduction": [],
    "Speed Improvement": []
}


accuracy_threshold_critical = 50
accuracy_threshold_good = 80
mistakes_threshold_critical = 4
mistakes_threshold_good = 1
speed_threshold = 90  # If speed is below 90, we suggest improvement


for index, row in df_performance.iterrows():
    topic = row['Topic']

    # Critical Improvement: Accuracy below 50% and mistakes above 4
    if row['accuracy'] < accuracy_threshold_critical or row['incorrect_answers'] > mistakes_threshold_critical:
        feedback["Critical Improvement"].append(f"Improve performance in '{topic}'. Focus on foundational understanding and practice more.")

    # Moderate Improvement: Accuracy between 50-79% and mistakes between 2-4
    elif row['accuracy'] >= accuracy_threshold_critical and row['accuracy'] < accuracy_threshold_good:
        feedback["Moderate Improvement"].append(
            f"'{topic}' shows moderate performance. Practice quizzes and review materials to strengthen understanding."
        )

    # Good Performance: Accuracy above 80% and few mistakes (less than 2)
    elif row['accuracy'] >= accuracy_threshold_good and row['incorrect_answers'] <= mistakes_threshold_good:
        feedback["Good Performance"].append(f"'{topic}' shows good performance. Continue maintaining consistency!")

    # Mistake Reduction: More than 2 mistakes
    if row['incorrect_answers'] > 2:
        feedback["Mistake Reduction"].append(
            f"Work on minimizing mistakes in '{topic}'. Focus on understanding why errors occur."
        )

    # Speed Improvement: Speed below 90 (assumed speed is in percentage)
    if row.get('speed', 100) < speed_threshold:  # Defaulting speed to 100 if not available
        feedback["Speed Improvement"].append(
            f"Consider working on time management in '{topic}'. Focus on answering questions faster during practice."
        )


for category, messages in feedback.items():
    print(f"\n{category}:")
    if messages:
        for message in messages:
            print(f"- {message}")
    else:
        print("No feedback available.")
```

# ***16. Personalized Recommendations***

```python
############## Define thresholds for score and accuracy######################
score_thresholds = {
    "excellent": 80,
    "good": 60,
    "needs_improvement": 0
}

accuracy_thresholds = {
    "excellent": 80,
    "good": 60,
    "needs_improvement": 0
}
```

```python
# Sample data for testing
score = 32  # Example score
accuracy = 80.0  # Example accuracy in percentage

# Function to evaluate performance and generate recommendations
def generate_recommendation(score, accuracy):
    recommendations = []

    # Score-based recommendations
    if score >= score_thresholds["excellent"]:
        recommendations.append("Great job! Keep up the good work.")
    elif score >= score_thresholds["good"]:
        recommendations.append("Good job! Keep practicing to improve.")
    else:
        recommendations.append("You can improve! Focus on the areas where you lost marks.")

    # Accuracy-based recommendations
    if accuracy >= accuracy_thresholds["excellent"]:
        recommendations.append("Excellent accuracy! Keep up the focus.")
    elif accuracy >= accuracy_thresholds["good"]:
        recommendations.append("Good accuracy, but aim for higher precision.")
    else:
        recommendations.append("Accuracy needs improvement. Review the mistakes.")

    return recommendations

# Test the function
recommendations = generate_recommendation(score, accuracy)
for recommendation in recommendations:
    print(recommendation)
```

```python
from reportlab.lib.pagesizes import letter
from reportlab.lib import colors
from reportlab.pdfgen import canvas

# Function to generate the PDF report
def generate_pdf_report(quiz_data, historical_data, recommendations, file_path="Quiz_Performance_Report.pdf"):
    pdf = canvas.Canvas(file_path, pagesize=letter)
    pdf.setTitle("Quiz Performance Report")

    # Add a title
    pdf.setFont("Helvetica-Bold", 16)
    pdf.drawString(100, 750, "Quiz Performance Report")

    # Add current quiz data
    pdf.setFont("Helvetica", 12)
    pdf.drawString(50, 720, "Current Quiz Data:")
    y_position = 700
    for key, value in quiz_data.items():
        pdf.drawString(70, y_position, f"{key}: {value}")
        y_position -= 20

    # Add historical data
    pdf.drawString(50, y_position, "Historical Data (Average):")
    y_position -= 20
    for key, value in historical_data.items():
        pdf.drawString(70, y_position, f"{key}: {value}")
        y_position -= 20

    # Add recommendations
    pdf.drawString(50, y_position, "Recommendations:")
    y_position -= 20
    for rec in recommendations:
        pdf.drawString(70, y_position, f"- {rec}")
        y_position -= 20

    # Save the PDF
    pdf.save()

# Call the function with necessary data
generate_pdf_report(
    quiz_data={
        "Score": 85,
        "Accuracy": "85%"
    },
    historical_data={
        "Average Score": 78,
        "Average Accuracy": "80%"
    },
    recommendations=[
        "Focus on improving time management.",
        "Review incorrect answers to identify patterns.",
        "Aim for consistent accuracy improvement."
    ]
)
```
